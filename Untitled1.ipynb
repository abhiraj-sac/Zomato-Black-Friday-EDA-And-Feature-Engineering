```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("zomato.csv",encoding='latin-1')
# df.head(10)
# df.columns
# df['City'].head(20)
# df.info()
# df.isnull().sum()
[feature for feature in df.columns if df[feature].isnull().sum() > 0]
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
# country_names = final_df['Country'].value_counts().index
# # Basically Zomato is centered in India and others are just a kinda outliers
# country_counts = final_df['Country'].value_counts().values
country_names=final_df.Country.value_counts().index
country_val=final_df.Country.value_counts().values
## Pie Chart- Top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
# plt.pie(country_counts,labels=country_names[:3])
final_df.columns
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns ={0:"size"})
```

```python
plt.figure(figsize=(10,6))
# sns.barplot(x="Aggregate rating",y="size",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
sns.countplot(data=ratings,x='Rating color',palette=['blue','red','orange','yellow','green','green'])
plt.show()
```
